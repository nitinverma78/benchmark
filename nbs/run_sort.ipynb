```python
#| export
from benchmark import bench
from fastcore.meta import delegates
from inspect import signature as sig
from fastcore.script import call_parse
```

```python
#| export
@delegates(bench.benchmark)
#def _main(target=1,**kwargs):
def _main(**kwargs):
    " customized main for sort " 
    return bench.benchmark(fldr='../sort', **kwargs)
    #return bench.benchmark(fldr='../search', **kwargs)
sig(_main)
#_main(data_sizes=[10, 1000], dist_families=["range","int"])
```

```python
#| export
@call_parse
def main(data_size:int=10, #data size
         dist_fam:str="range", #data distributions
         num_variations:int=1, #data variations
         )->None:
    "plot run-times for sort algorithms"
    return _main(data_sizes=[data_size], dist_families=[dist_fam])
```

```python
show_doc(main)
```

```python
main()
```
